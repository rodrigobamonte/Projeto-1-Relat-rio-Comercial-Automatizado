# Projeto 1 — Relatório Comercial Automatizado

> Objetivo:

Desenvolver um relatório comercial automatizado utilizando Python para calcular indicadores de desempenho e gerar um resumo executivo a partir de dados de uma empresa fictícia.

> Contexto:

A Prime Travel Consulting é uma empresa fictícia especializada em consultoria para documentação internacional e atua por meio de uma rede de agências parceiras espalhadas pelo Brasil.

Ao final de cada mês, a equipe comercial precisa analisar os principais indicadores da operação para acompanhar o desempenho das agências e apoiar a tomada de decisões.

Neste projeto, o objetivo é automatizar esses cálculos utilizando Python.

**DADOS DO MÊS**
| Indicador                            |            Valor |
| ------------------------------------ | ---------------: |
| Total de Agências Parceiras          |           **80** |
| Agências Ativas (indicaram clientes) |           **26** |
| Agências com Venda                   |           **17** |
| Total de Leads                       |          **156** |
| Total de Vendas                      |           **63** |
| Faturamento                          | **R$ 28.490,75** |

Ao longo do desenvolvimento do projeto, serão calculados os seguintes indicadores comerciais:

* Taxa de Ativação das Agências – percentual de agências que realizaram indicações no período.

* Percentual de Agências com Venda – percentual de agências que efetivamente geraram vendas.

* Taxa de Conversão – percentual de leads convertidos em vendas.

* Ticket Médio – valor médio gerado por cada venda realizada.

* Leads Não Convertidos – quantidade de leads que não resultaram em vendas.

* Receita Média por Agência com Venda – valor médio de faturamento gerado por cada agência que teve uma venda.

# **Taxa de Ativação das Agências**

A Taxa de Ativação indica, em percentual, quantas agências parceiras realizaram pelo menos uma indicação de cliente durante o período analisado.

Neste projeto, considera-se uma agência ativa aquela que realizou ao menos uma indicação.

**Fórmula:**
Agências Ativas / Total de Agências Parceiras * 100

```python
# Taxa de Ativação

agencias_parceiras = 80
agencias_ativas = 26
meta_ativacao = 30

taxa_ativacao = agencias_ativas / agencias_parceiras * 100
print(f"A taxa de ativação das agências é de {taxa_ativacao}%")

if taxa_ativacao >= meta_ativacao:
    print("Meta de ativação de agência atingida")
else:
    print("Meta de ativação de agência não atingida")
```

# **Percentual de Agências com Venda**

O Percentual de Agências com Venda representa a porcentagem de agências parceiras que efetivamente geraram pelo menos uma venda no período.

Esse indicador permite avaliar quantas agências, além de indicar clientes, conseguiram gerar resultados concretos para a empresa.

***Fórmula:**
Agências com Venda / Total de Agências Parceiras * 100

```python
# Percentual de Agências com Venda

agencias_parceiras = 80
agencias_vendas = 17
meta_percentual = 30

perc_agencias_venda = agencias_vendas / agencias_parceiras * 100
print(f"O percentual de agências com venda é de {perc_agencias_venda}%")

if perc_agencias_venda >= meta_percentual:
    print("Meta de venda de agência atingida")
else:
    print("Meta de venda de agência não atingida")
```

# **Taxa de Conversão**

A Taxa de Conversão representa o percentual de leads indicados que foram convertidos em vendas.

Esse indicador permite avaliar a eficiência do processo comercial, além de fornecer informações sobre a qualidade dos leads enviados pelas agências parceiras.

A partir dessa análise, é possível identificar oportunidades de melhoria nas abordagens comerciais, compreender o perfil dos clientes e localizar possíveis gargalos ao longo do funil de vendas.

**Fórmula:**
Total de Vendas / Total de Leads * 100

```python
# Taxa de Conversão

total_leads = 156
total_vendas = 63
meta_conversao = 25

taxa_conversao = total_vendas / total_leads * 100
print(f"A taxa de conversão é de {taxa_conversao}%")

if taxa_conversao >= meta_conversao:
    print("Meta de conversão atingida")
else:
    print("Meta de conversão não atingida")
```

# **Ticket Médio**

O Ticket Médio representa o valor médio gerado por cada venda realizada durante o período.

Esse indicador permite compreender quanto, em média, cada venda contribuiu para o faturamento da empresa.

**Fórmula:**
Faturamento / Total de Vendas

```python
# Tiktet Médio

faturamento = 28490.75
total_vendas = 63

tiktet_medio = faturamento / total_vendas
print(f"O ticket médio é de {tiktet_medio}")
```

# **Leads Não Convertidos**

O indicador de Leads Não Convertidos representa a quantidade de leads que não resultaram em vendas.

Sua análise permite identificar possíveis perdas ao longo do processo comercial e investigar se essas perdas estão relacionadas à abordagem da equipe de vendas, ao perfil dos clientes ou à qualidade das indicações realizadas pelas agências parceiras.

**Fórmula (quantidade):**
Total de Leads - Total de Vendas

**Fórmula (percentual):**
Leads Não Convertidos / Total de Leads * 100

```python
# Leads não convertidos

total_leads = 156
total_vendas = 63
meta_leads_nao_convertidos = 40

leads_nao_convertidos = total_leads - total_vendas
percentual_leads_nao_convertidos = (leads_nao_convertidos / total_leads) * 100
print(f"O percentual de leads não convertidos é de {percentual_leads_nao_convertidos}%")

if percentual_leads_nao_convertidos >= meta_leads_nao_convertidos:
    print("Meta de leads não convertidos atingida")
else:
    print("Meta de leads não convertidos não atingida")
```

# **Receita Média por Agência com Venda**

A Receita Média por Agência com Venda representa o faturamento médio gerado por cada agência parceira que realizou pelo menos uma venda durante o período analisado.

Esse indicador auxilia na avaliação da produtividade das agências que efetivamente contribuíram para o faturamento da empresa.

**Fórmula:**
Faturamento / Agências com Venda

```python
# Receita Média por Agência com Venda

faturamento = 28490.75
agencias_vendas = 17

receita_media_por_venda = faturamento / agencias_vendas
print(f"A receita média por venda é de {receita_media_por_venda}")
```
